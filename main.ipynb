## 1. 项目概述

### 1.1 产品名称与功能：

**名称：**小电bot（Web项目）   
**功能：**虚拟对话机器人，sqlite（+Chroma）电力市场信息数据库，文件自动摘要助手  
**简介：**基于通义千问+RAG助力电力市场与电力交易有关文件摘要与智能问答系统，面向政务人员、企业与个人分别提供定制化服务   


### 1.2 项目背景：

2024年7月2日，南方区域电力市场管理委员会在广州成立。这是全国首个区域级电力市场管理委员会，标志着南方区域电力市场向打造全国统一电力市场体系建设标杆迈出重要一步。国家能源局高度重视管委会组建工作，要求管委会发挥专业优势，加强沟通协同，汇聚各方力量，共同推进南方区域电力市场建设，为全国统一电力市场体系提供样板、作出示范。与此同时，其他省市如福建、浙江也在加速建立相关机构，大量的相关管理规范文件需要商议、形成、并且得到有效的推广。

为此，我设计了一个模型基于通义千问LLM，数据增强依赖sqlite和Chroma的对话机器人小电bot，能够综合外部数据源和检索增强数据源，定向地帮助电力市场这一物理系统中的参与者快速摘要相关文件，领会文件内容；系统提供了智能问答的功能，方便用户咨询相关问题；同时提供了写入数据库的接口，方便进一步增加和强化内部数据源，提高系统的专用性。



### 1.3 项目动机：

- 我的动机是降低群众、企业与政府工作人员理解电力市场文件的门槛，同时减少简单问题咨询的成本、提升规章制度推广的效率。传统模式下，政府下发的电力市场文件数量多而且缺少管理，我认为很难满足新时代的电力市场成员和电力市场监管委员会的需要；


- 我希望通过引入大语言模型+检索增强的技术，为电力市场的发电方、输电方、购电方、市场运营机构、第三方等成员提供一个通用的文件摘要与智能对话系统，增加企业和群众对于电力市场组织机构和工作、议事规则的了解，减少系统中成员对于有关管理规范文件的困惑。


- 我希望通过引进该模型，可以通过区分在市场中的不同权能与角色来提供个性化的功能，可以对不同类型的用户使用情况的反馈对用户画像，为电力市场从业人员提供相较于传统方式更有针对性的个性化信息检索与分析指导。


- 同时对于个人而言，这也是我最近一段时间以来想尝试和正在尝试的方向，即如何通过RAG构建比较专用的大模型，以专用大模型赋能专门方向的产业。我觉得我的方向是有价值的。🤗



### 1.4 重要性与价值：

- 目前电力市场的开放程度越来越高，售电公司、用户等市场主体可以直接参与南方区域电力市场的议事协调，储能企业也首次成为南方区域市场管委会代表。目前，南方区域电力市场注册主体已达到18万家，同比增长19.1%，其中发电企业1169家、售电公司709家、电力用户17.89万家。中长期省内市场化电量由2016年的1519亿千瓦时增长到2023年的6213.7亿千瓦时，2024年5月已达到2666.1亿千瓦时，年均增速22.3%，市场开放程度走在全国前列。

- 本项目旨在助力电力市场与电力市场监管部门向智能化、个性化方向迈进。近年来，南方区域电力市场以打造“南方样板”为目标，加速推进市场建设，并已开展按周全域结算试运行。“南方样板”既是全国统一电力市场体系的重要实践，也是推动构建新型电力系统和新型能源体系的积极探索。而我的技术正式为了赋能全国统一电力市场体系，为电力市场建设提速的抓手。


- 在技术层面，小电bot通过引入计算机视觉、自然语言处理、检索增强的技术，有望帮助不断扩大的市场主体中的参与人员把握市场方向。


- 本项目不仅仅是一个智慧政务创新的尝试，更是在人工智能和政务服务融合的大潮中，为构建更高效、更智能、更公平的电力市场生态系统而做出的一次积极探索。🧐





## 2. 问题陈述


### 2.1 问题描述

**1.小电bot将尝试主要解决以下问题：**

- 政务文件繁杂：智能摘要帮助电力市场从业人员和参与者生成相关文件内容的关键信息摘要，减少阅读和整理时间。
- 相关从业人员文件整理工作繁重：智能数据库帮助存放、整理和检索相关文件内容，提高工作效率，减少重复劳动。
- 读图效率提升：大模型帮助分析图片内容，提供图片主旨，提高视觉信息的处理效率。
- 实时政策更新：及时获取并解析最新的电力市场政策，确保从业人员和参与者能够快速掌握最新动态。
- 市场分析复杂：提供智能化的市场分析报告，帮助从业人员做出更加明智的决策。
- 数据海量难以处理：利用大模型的强大处理能力，快速分析海量数据，提取关键信息。

**2.解决以上问题的思路与可能的难点：**

- 系统设计的关键是利用大模型的强大阅读、推理和生成能力。通过规范且系统的prompt工程，结合强大的向量数据库，为用户提供专业的内容生成服务。对于文本摘要和分析，需要利用大模型自动生成文件摘要和分析报告，提高文本处理效率。通过大模型的图像识别功能，分析图像内容，提取关键信息。

- 实时更新和通知：实时获取并解析最新政策和市场动态，及时通知相关人员。

- 我着重考虑了电力系统专用性模块的构建。将“适应性学习”的方法用prompt工程在系统内进行复刻，为电力市场从业人员提供相较于传统方式更有针对性的个性化信息检索与分析指导，这是本项目的核心点也是创新点之一。根据用户的行为和需求，智能推送相关信息和报告，实现个性化内容推送。

- 本项目的可能的难点主要来自内容生成：对于常规对话场景或向量数据库体数据量较小时，大模型可能会生成不准确或与实际内容不符的信息。对于引导类内容生成的场景下，如果系统提示词不够完备，可能会跳过引导内容直接生成结论内容，影响信息的准确性和实用性。

- 除此之外，在处理政务文件和市场分析数据时，必须确保数据隐私和安全，防止数据泄露和滥用。同时需要确保系统能够适应不同规模和复杂度的电力市场环境，提供高效的内容生成和数据分析服务。为此，系统架构需要具备良好的扩展性和灵活性。另外，针对巨大的用户群体，需要设计直观易用的用户界面，提供便捷的交互方式，并建立有效的用户反馈机制，不断改进和优化系统性能和功能。

### 2.2 项目目标

**本次项目的目标与预期成果：**

- **功能实现：**实现功能架构中80%以上的的功能点，实现的功能点达到90%以上的完整性
- **可靠性：**通过prompt工程对模型幻觉难点进行调优，内容准确率达到95%以上
- **用户体验：**通过合理的界面与操作逻辑的构建，实现简洁大方且交互逻辑合理的系统界面
- **扩展性：**确保系统架构具备良好的扩展性，能够适应不同规模和复杂度的电力市场环境。

**本项目从底向上依次分为 LLM 层、数据层、数据库层、应用层与服务层：**


1. **LLM 层：** 在这一层，我们采用通义千问大模型API进行调用封装。这层提供了统一的入口和方式，

2. **数据处理层：** 数据层包括个人知识库的源数据和Embedding API。源数据通过Embedding处理后，可以被向量数据库使用。同时使用外部的工具接口进行数据处理的辅助。

3. **数据库层：** 这一层主要构建在个人知识库源数据之上的向量数据库，本项目将选择Chroma作为数据库；同时使用了sqlite作为系统的数据库。

4. **应用层：** 应用层是核心功能的最顶层封装。我们在这里使用LangChain构建基于通义千问大模型的问答链基类以便实现基于数据库的检索问答，同时为保证输出内容针对性与质量我们加上了系统prompt工程。

5. **服务层：** 最顶层是服务层，主要是基于应用层的核心功能封装，实现了 Demo 的搭建或 API 的封装。在本项目中，我分别实现了通过 Vue 3.0（Vite） 搭建前端界面与 flask 进行后端接口封装（暂定），支持多样化的项目调用。

   

## 4. 应用场景


**（这一部分我们用三段虚构小故事来体现）**

- 故事一

小Mo是居住在N市的公民，虽然平时有条件网上缴纳电费，但他还是对电力市场收费问题比较关注。
这个月，他新买入了显卡和电脑，不知不觉在用电高峰时间使用了大量电力。
月末他收到电费账单时，为激增的电费大吃一惊。
听说24年7月N市刚刚成立了一个电力市场监管委员会，小Mo打算咨询一下电费为何突然激增。
为了一探究竟，他下载了N市电力市场监管委员会最新推出的“小电bot”

进入电力管理/数据分析界面，他清晰地看到了自己在高峰时期的大量耗电；
同时有实时的现价电费显示在界面上。
他这才了解到不同时间的用电会有不同的收费标准。

“不同时间用不一样的收费标准，这不是骗老百姓的钱嘛”，怀着不理解的心情，他打开了小电bot智能问答助手。
小Mo向助手提问：“为什么不同时间电费的收费标准不同？”
对话框中小电bot开始解释，不同时间用电对电网的压力不同，加上电力难以储存的特点，采取峰时收费高谷时收费低的标准，同时罗列了N市的收费标准文件。
小Mo没有耐心查看文件，于是要小电bot截取文件中的关键内容，为小Mo生成了文件的摘要。

通过和小电bot的智能助手交流和平台上的搜索学习，小Mo对于电力市场和用电的知识增加了许多。
在个人推荐模块，小Mo了解了更加多他感兴趣的用电知识和省电小技巧。
小Mo还专门办了一个小型储电器和用电卡，在低价电时储电，在高峰期卖电。
小Mo通过储电器小赚了一点，N市的储电压力和峰时用电压力也因为小Mo这样的用户得到了一定的缓解。
现在，小Mo打算全款买下《黑神话·悟空》，以他对电力市场的了解和存足了的电力卡，
即使是开全性能玩游戏，他也有把握较上个月节约下一大笔电费了。

- 故事二

小Mo在N市有一家玻璃工厂。日夜不停的工厂为他带来了巨大的利润，但巨额的电力收费也随之而来。
小Mo时常好奇：电力市场到底是如何收费和运营的，如果能减少电费，工厂的效益能再上一层。
2024年7月，N市成立电力市场监管委员会，除了政务人员外，还邀请了电力企业、用电企业、普通居民等作为第三方监管，小Mo也在受邀的行列内。
为了搞明白电力市场监管委员会的运作方法，同时减少自家电费的消耗，小Mo下载了最新推出的小电bot。

打开小电bot，他最先查看的是自家电力的使用情况，
安装了小米插座之后，电力消耗可以追踪到工位。
他发现三车间某一工位的用电明显异常，
调查以后发现是设备接触不良，导致了此处电力的大量浪费。

小Mo又打开了风险管理，发现三车间那一工位属于高风险。
他赶紧要求设备方检修更换了设备，一场酝酿中的危机也因此消弭于无形。

小电bot同时为他提供了电力项目管理的功能，小Mo在小电bot上规划用电项目，争取在谷时进行高耗电的项目，
查看小电bot提供的用电效率报告，小Mo的工厂效益果然得到了提高。

除此之外，小Mo还通过小电bot的政务文件处理、智能问答、信息检索等功能了解了电力市场运营的规范，更好地肩负起监管方的职责……

- 故事三

小Mo是N市电力系统中的一名政务人员，他时长苦恼，群众对于用电安全、收费标准等还没有建立基本的认知。
大量的基层服务人员被简单重复的任务困扰，而许多业务常常排队到下班也无法解决。
近来N市成立电力市场监管委员会，由小Mo牵头领导，推出了“小电bot”，旨在解决民众一般的咨询问题，提高电力市场政策的宣传推广能力，同时及时收集群众和专家意见。

小电bot开发完成后，小Mo第一个用上了小电bot。
小Mo发现小电bot对政务人员也有极大的帮助，它摘要文件、解读文件的能力不错，许多冗长而非重要的文件可以交给它来整理。
基于sqlite的数据库还能及时保存和更新内部的模型数据源。
同时，小电bot还可以侦测城市各个区域的用电热力图，对于电力系统调配电力，检查企业和个人偷电用电，约谈用电效率低的企业都有了工作的依据……

### 6.2 需求分析


#### 6.2.1 面向政务群体
**1.1 政务群体在电力市场管理中的痛点与需求点分析：**

电力市场的管理涉及到多个方面的复杂事务，包括政策制定、市场监管、用户反馈处理等。政务人员在日常工作中会面临以下痛点和需求：

````python
"""
Author: linyifan
Time: 2024年7月10日10:31:02
功能:
- 加载文档、拆分文档并初始化 Chroma 数据库。
- 实现知识库管理类和文本生成类，用于处理不同类型的文件并生成文本。
"""
import sqlite3
import json
import os
import requests
from langchain_community.llms import Tongyi
import PyPDF2
import docx
from PIL import Image
import pytesseract
from bs4 import BeautifulSoup
from langchain.document_loaders import TextLoader
from langchain.text_splitter import CharacterTextSplitter
from langchain.vectorstores import Chroma
from langchain.embeddings import OpenAIEmbeddings

os.environ["OPENAI_API_KEY"] = ""

# 加载文档
file_path = "/home/jovyan/work/项目2-大模型应用创意设计/temp.md"
docs = TextLoader(file_path).load()

# 拆分文档
text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=0)
documents = text_splitter.split_documents(docs)

# 初始化 Chroma 数据库
chroma_db = Chroma.from_documents(documents, OpenAIEmbeddings())

# 设置通义千问 API 密钥
os.environ["DASHSCOPE_API_KEY"] = ""

class KnowledgeBase:
    """
    知识库管理类，负责与 SQLite 数据库的交互，并处理 Chroma 数据库。
    """
    
    def __init__(self, db_path, chroma_db):
        """
        初始化方法。

        参数:
        db_path (str): SQLite 数据库文件路径。
        chroma_db (Chroma): 初始化的 Chroma 数据库对象。
        """
        self.db_path = db_path
        self.chroma_db = chroma_db

    @property
    def connection(self):
         """
        创建并返回 SQLite 数据库连接。

        返回:
        sqlite3.Connection: SQLite 数据库连接对象。
        """
        return sqlite3.connect(self.db_path)

    def retrieve_chroma_info(self, query):
        """
        从 Chroma 数据库中检索与查询相关的信息。

        参数:
        query (str): 查询字符串。

        返回:
        list: 检索到的相关信息及其来源。
        """
        results = self.chroma_db.similarity_search(query, k=5)
        return [(res.metadata['source'], res.page_content) for res in results]

    def retrieve_info(self, query):
        """
        从 SQLite 数据库和 Chroma 数据库中检索与查询相关的信息。

        参数:
        query (str): 查询字符串。

        返回:
        tuple: 内部信息和 Chroma 检索结果。
        """
        conn = self.connection
        cursor = conn.cursor()

        cursor.execute("SELECT title, content FROM internal_info WHERE content LIKE ?", ('%' + query + '%',))
        internal_results = cursor.fetchall()

        chroma_results = self.retrieve_chroma_info(query)

        conn.close()

        return internal_results, chroma_results

    def add_internal_info(self, source, content):
        """
        向 SQLite 数据库中添加内部信息。

        参数:
        title (str): 信息标题。
        content (str): 信息内容。
        """
        conn = self.connection
        cursor = conn.cursor()

        cursor.execute("INSERT INTO external_info (source, content) VALUES (?, ?)", (title, content))

        conn.commit()
        conn.close()
        
    def add_external_info(self, title, content):
        """
        向 SQLite 数据库中添加外部信息。

        参数:
        title (str): 信息标题。
        content (str): 信息内容。
        """
        conn = self.connection
        cursor = conn.cursor()

        cursor.execute("INSERT INTO internal_info (title, content) VALUES (?, ?)", (title, content))

        conn.commit()
        conn.close()
        
    def add_chroma_info(self, source, content):
        """
        向 Chroma 数据库中添加外部信息。

        参数:
        source (str): 信息来源。
        content (str): 信息内容。
        """
        documents = text_splitter.split_documents([TextLoader.from_string(content)])
        self.chroma_db.add_documents(documents, metadata=[{"source": source}])

class TextGenerator:
    """
    文本生成类，负责根据查询生成文本，并处理不同类型的文件。
    """
    def __init__(self, knowledge_base, temperature=0.7):
        """
        初始化方法。

        参数:
        knowledge_base (KnowledgeBase): 知识库管理对象。
        temperature (float): 生成文本的温度参数。
        """
        self.knowledge_base = knowledge_base
        self.llm_tongyi = Tongyi()
        self.temperature = temperature

    def generate_text(self, query):
        """
        根据查询生成文本。

        参数:
        query (str): 查询字符串。

        返回:
        str: 生成的文本。
        """
        internal_results, chroma_results = self.knowledge_base.retrieve_info(query)

        context = "内部信息:\n"
        for title, content in internal_results:
            context += f"标题: {title}\n内容: {content}\n\n"

        context += "外部信息:\n"
        for source, content in chroma_results:
            context += f"来源: {source}\n内容: {content}\n\n"

        # 使用通义千问API进行文本生成
        prompt = f"{context}\n{query}"
        generated_text = self.llm_tongyi.invoke(prompt, temperature=self.temperature)

        return generated_text

    def process_file(self, file_path):
        """
        处理不同类型的文件并提取文本。

        参数:
        file_path (str): 文件路径。

        返回:
        str: 提取的文本内容。
        """
        _, file_extension = os.path.splitext(file_path)
        if file_extension == ".json":
            with open(file_path, 'r', encoding='utf-8') as file:
                data = json.load(file)
            return json.dumps(data, ensure_ascii=False)
        elif file_extension == ".pdf":
            return self._process_pdf(file_path)
        elif file_extension == ".docx":
            return self._process_docx(file_path)
        elif file_extension in [".jpg", ".jpeg", ".png"]:
            return self._process_image(file_path)
        else:
            raise ValueError("Unsupported file type")

    def _process_pdf(self, file_path):
        """
        处理PDF文件并提取文本。

        参数:
        file_path (str): PDF文件路径。

        返回:
        str: 提取的文本内容。
        """
        text = ""
        with open(file_path, 'rb') as file:
            reader = PyPDF2.PdfReader(file)
            for page_num in range(len(reader.pages)):
                page = reader.pages[page_num]
                text += page.extract_text()
        return text

    def _process_docx(self, file_path):
        """
        处理DOCX文件并提取文本。

        参数:
        file_path (str): DOCX文件路径。

        返回:
        str: 提取的文本内容。
        """
        doc = docx.Document(file_path)
        text = "\n".join([para.text for para in doc.paragraphs])
        return text

    def _process_image(self, file_path):
        """
        处理图像文件并提取文本。

        参数:
        file_path (str): 图像文件路径。

        返回:
        str: 提取的文本内容。
        """
        image = Image.open(file_path)
        text = pytesseract.image_to_string(image)
        return text

    def process_webpage(self, url):
        """
        处理网页并提取文本。

        参数:
        url (str): 网页URL。

        返回:
        str: 提取的网页文本内容。
        """
        response = requests.get(url)
        soup = BeautifulSoup(response.content, 'html.parser')
        text = soup.get_text()
        return text

# 使用示例
db_path = 'knowledge_base.db'
kb = KnowledgeBase(db_path, chroma_db)
tg = TextGenerator(kb)

query = "电力市场的主要参与者有哪些？"
generated_text = tg.generate_text(query)
print(generated_text)

file_text = tg.process_file('/home/jovyan/work/项目2-大模型应用创意设计/福建省电力管理委员会章程.pdf')
query = "总结以下文本，用```来标识开始和结束：```" + "上述内容无需关注，一下为需要总结的文本：\n" + file_text + "```"
generated_text = tg.generate_text(query)
print(generated_text)

webpage_text = tg.process_webpage('https://mp.weixin.qq.com/s/MVgVJvBCJ4encsokZiF6uA')
query = "总结以下文本，用```来标识开始和结束：```" + "上述内容无需关注，一下为需要总结的文本：\n" + webpage_text + "```"
# 经查验webpage_text能成功读到网页内容，那么与file_text应当同理可以作为tg.generate_text()的输入，考虑到token量不做进一步展示。
# generated_text = tg.generate_text(query)
# print(generated_text)
````

### 后端部署（使用Flask框架）

````python
"""
Author: linyifan
Time: 2024年7月10日18:00:02
功能:
- 后端环境搭建
"""
from flask import Flask, request, jsonify

app = Flask(__name__)

# 初始化知识库和文本生成器
db_path = 'knowledge_base.db'
kb = KnowledgeBase(db_path)
tg = TextGenerator(kb)

@app.route('/generate', methods=['POST'])
def generate():
    """
    处理文本生成请求。

    参数:
    - query (str): 查询字符串。
    - temperature (float): 生成文本的温度参数。

    返回:
    - json: 生成的文本或错误信息。
    """
    data = request.get_json()
    query = data.get('query', '')
    temperature = data.get('temperature', 0.7)

    if not query:
        return jsonify({"error": "Query is required"}), 400

    tg.temperature = temperature
    generated_text = tg.generate_text(query)
    return jsonify({"generated_text": generated_text})

@app.route('/process_file', methods=['POST'])
def process_file():
    """
    处理文件上传并提取文本。

    参数:
    - file: 上传的文件。

    返回:
    - json: 提取的文本或错误信息。
    """
    if 'file' not in request.files:
        return jsonify({"error": "No file part"}), 400
    
    file = request.files['file']
    file_path = os.path.join('/tmp', file.filename)
    file.save(file_path)

    try:
        file_text = tg.process_file(file_path)
        return jsonify({"file_text": file_text})
    except ValueError as e:
        return jsonify({"error": str(e)}), 400

@app.route('/add_internal', methods=['POST'])
def add_internal():
    """
    添加内部信息到知识库。

    参数:
    - title (str): 信息标题。
    - content (str): 信息内容。

    返回:
    - json: 成功添加的消息或错误信息。
    """
    data = request.get_json()
    title = data.get('title')
    content = data.get('content')
    
    if not title or not content:
        return jsonify({'error': 'Title and content are required'}), 400
    
    kb.add_internal_info(title, content)
    return jsonify({'message': 'Internal info added successfully'}), 201

@app.route('/add_external', methods=['POST'])
def add_external():
    """
    添加外部信息到知识库。

    参数:
    - source (str): 信息来源。
    - content (str): 信息内容。

    返回:
    - json: 成功添加的消息或错误信息。
    """
    data = request.get_json()
    source = data.get('source')
    content = data.get('content')
    
    if not source or not content:
        return jsonify({'error': 'Source and content are required'}), 400
    
    kb.add_external_info(source, content)
    return jsonify({'message': 'External info added successfully'}), 201

@app.route('/add_chroma', methods=['POST'])
def add_chroma():
    """
    添加信息到 Chroma 数据库。

    参数:
    - source (str): 信息来源。
    - content (str): 信息内容。

    返回:
    - json: 成功添加的消息或错误信息。
    """
    data = request.get_json()
    source = data.get('source')
    content = data.get('content')
    
    if not source or not content:
        return jsonify({'error': 'Source and content are required'}), 400
    
    kb.add_chroma_info(source, content)
    return jsonify({'message': 'Chroma info added successfully'}), 201

if __name__ == '__main__':
    app.run(debug=True,port=5009)
````
